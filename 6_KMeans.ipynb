# K MEANS CLUSTERING ALGORITHM

```python
# K-means is an unsupervised clustering algorithm 
# Partition unlabelled data into a certain number (thats the “ K”) of distinct groupings. 
# classifies them together into clusters.
# Prediction is not done under this technique
```

```python
# Creates the cluster / groups based on the similarity 
# similar data points become one cluster/ different data points forms the another cluster
# similarity is based on the shortest diatance between the points(eucledian distance )
# similar kind of data points become one cluster, 
# randomly choose an initial centroid (centre coordinates) for each cluster

# strating point is ramdomly chosen centroid. 
# i cant stop my model after first iteration because the initial points are random. not a proper model
# Need to run this model number of times to get the better model 
# but iteration would not be a random point but a claculated point...
# New mean calculated for each cluster and that point will be your new centroid 
# repeat the process of creating clusters 

# When using the calculated value, the centroid slightly changes - Repeat the process
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sns = pd.read_csv("snsdata.csv")
sns.shape
```

```python
sns.head()
```

```python
# Data Description 
# The SNS dataset contains 30000 observations (rows) 
# each represents a high school student and 40 features (columns) 
# that provides information for the student. 

# 1 - Gradyear: Graduation year of the student (2006, 2007, 2008, 2009) 
# 2 - Gender: Gender of the student (male, female) 
# 3 - Age: Age of the student 
# 4 - Friends: No of friends 
```

```python
sns.gender.replace({"M" :1 , "F" :0} , inplace = True)
```

```python
sns.isnull().sum()
```

```python
sns.gender  = sns.gender.fillna(1)
sns.age     = sns.age.fillna(sns.age.mean())
```

```python
sns.isnull().sum()
```

```python
sns.head()
```

```python
from sklearn.cluster import KMeans
```

```python
sns_kmeans = KMeans(n_clusters=4) 
#n_clusters=4 => 0,1,2,3
```

```python
#here training and testing is not done because it a unsupervised learning algorithm.
sns_kmeans.fit(sns)
```

```python
sns_kmeans.labels_
```

```python
#to see all
list(sns_kmeans.labels_)
```

```python
len(list(sns_kmeans.labels_))
```

```python
sns_kmeans.cluster_centers_
```

```python
pd.DataFrame(sns_kmeans.cluster_centers_)
```

```python
sns_kmeans.fit(sns).score(sns)  # gives the value of SSD for each value of K
```

ELBOW METHOD 

```python
nc = range(1 ,8)  
kmeans = [ KMeans(n_clusters = i ) for i in nc]
kmeans
score = [kmeans[i].fit(sns).score(sns) for i in range(len(kmeans)) ]
print(score) 
```

```python
x=list()
for i in range(1,10):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(sns)
    y=kmeans.fit(sns).score(sns)
    x.append(y)
```
